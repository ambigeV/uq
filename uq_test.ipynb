```python
import deepchem as dc
from deepchem.molnet import load_qm7, load_delaney, load_freesolv, load_pdbbind
from sklearn.svm import SVR
```

```python
# tasks, datasets, transformers = load_delaney(featurizer='ECFP')
tasks, datasets, transformers = load_qm7()
train, valid, test = datasets
```

```python
sk_model = SVR()
dc_model = dc.models.SklearnModel(sk_model)

# 4. Fit on the DeepChem NumpyDataset
dc_model.fit(train)

# 5. Evaluate
metric = dc.metrics.Metric(dc.metrics.mean_squared_error)
valid_scores = dc_model.evaluate(valid, [metric])
test_scores = dc_model.evaluate(test, [metric])

print("Validation MSE:", valid_scores[metric.name])
print("Test MSE:", test_scores[metric.name])
```
